# Remove and handls Outliers

### I dentify Outliers (visual)

```python
# Identify living area outliers
plt.figure()
sns.boxplot(df["sqft_living"])
plt.title("Identify living area Outliers")
plt.show()
```

```python
# Identify bedrooms outliers
plt.figure()
sns.boxplot(x=df["bedrooms"])
plt.title("Identify bedrooms outliers")
plt.show()
```

### Define IQR function

```python
def remove_outliers(data,column):
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3-1.5*IQR 
    return data[(data[column]>=lower) & (data[column]<=upper)]

```

## applying cleaning

```python
df=remove_outliers(df,"bedrooms")
```

```python
df=remove_outliers(df,"sqft_living")
```

```python
df=remove_outliers(df,"bathrooms")
```

### Price handl carefully

```python
price_upper=df["price"].quantile(0.99)
df["price"]=np.where (df["price"]>price_upper,price_upper,df["price"])
```
