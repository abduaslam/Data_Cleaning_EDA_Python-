# Data cleaning

## Check if there any missing or duplicate values

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.duplicated().sum()
```

### Rname some columns name

```python
df.columns
```

```python
df.rename(columns={"Publishing Year":"Publishing_Year","Book Name":"Book_Name","gross sales":"Gross_Sales","publisher revenue":"Publishe_Revenue",
                  "sale price":"Sales_Price","sales rank":"Sales_Rank","units sold":"Units_Sold"},inplace=True)
```

```python
df.columns
```

## Correct columns data types

```python
df.dtypes
```

```python
# change data type in  Puplishong_year  to date time
df["Publishing_Year"]=df["Publishing_Year"].astype('int64')
```

### Standardize languages code

```python
df["language_code"].unique()
```

```python
df["language_code"]=df["language_code"].replace({
    "en-US":"eng","en-GB":"eng"
    
})
```

#### clean genre column

```python
df["genre"]=df["genre"].str.lower().str.strip()

df["genre"]=df["genre"].replace({
    "gener fiction":"gener"})
```

### handle outlierds(unit sold & sales price

```python
df=df[df["Sales_Price"]>0]
df=df[df["Units_Sold"]>0]
```
