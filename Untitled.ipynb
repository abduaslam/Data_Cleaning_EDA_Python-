### Outlier Detections and Removal

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Salary"])
plt.title("Check salary outliers")
plt.show()            
```

```python
###  Remove outliers IQR (Intrequntile)mothed
```

```python
Q1=df['Salary'].quantile(0.5)
Q3=df['Salary'].quantile(0.75)
IQR=Q3-Q1
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
```

```python
  df=df[(df["Salary"]>=lower_limit)&(df["Salary"]<=upper_limit)]
```

```python
# Check Outlier again
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Salary"])
plt.title("Check salary outliers")
plt.show()            
```
