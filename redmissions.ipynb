## If there are missing values I will hanle with these mothed

```python
## numerical values
# select all numrical columns]
num_cols=df.select_dtypes(include=["int64","float64"]).columns
#fill out all numerical  values with median\
df[num_cols]=df[num_cols].fillna(df[num_cols].median())
```

```python
## This for categorical columns
#select all categrical columns
cate_cols=df.select_dtypes(include="object").columns
# fill missing values with "unknown"

df[cate_cols]=df[cate_cols].fillna("unknown")
```
